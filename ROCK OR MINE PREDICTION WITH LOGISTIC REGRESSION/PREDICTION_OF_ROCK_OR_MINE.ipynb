**SONAR ROCK VS MINE PREDICTION USING SONAR DATA**


> WE HAVE TO PREDICT WHETHER THE OBJECT IS AN ROCK OR MINE



*IMPORTING THE REQUIRED LIBRARIES*

```python
import pandas as pd
import numpy as np    # used to create numpy array
from sklearn.model_selection import train_test_split  # used to split the data into train and test
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

*DATA COLLECTION AND DATA PREPROCESSING*

```python
# HERE WE ARE LOADING THE DATASET INTO THE DATAFRAME (df)
df = pd.read_csv("/content/Copy of sonar data.csv",header=None) # In our dataset there are no header name due to that we are keeping the header value as NONE
df.head()
```

```python
# TO FIND THE NUMBER OF ROWS AND COLUMNS OF THE DATASET
df.shape
```

```python
# IT GIVES THE STATISTICAL MEASURES OF THE DATA
df.describe()
```

```python
df[60].value_counts()
```

M        MINE

R        ROCK

```python
df.groupby(60).mean()
```

```python
# SEPARATING DATA AND LABELS
x = df.drop(columns=60,axis=1)
y = df[60]
```

```python
x.head()
```

```python
y.head()
```

TRAINING AND TEST DATA

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1) # RANDOM STATE IS TO SPLIT THE DATA IN SAME MANNER
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
print(y_train)
```

TRAINING A MODEL


> LOGISTIC REGRESSION



```python
model1 = LogisticRegression()
```

```python
# TRAIN THE LOGISTIC REGRESSION MODEL  WITH THE TRAIN DATA
model1.fit(x_train,y_train)
```

MODEL EVALUATION

```python
# ACCURACY ON THE TRAINING DATA
x_train_prediction = model1.predict(x_train)
train_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("ACCURACY SCORE ON TRAINING DATA : ",train_data_accuracy)
```

```python
# ACCURACY ON THE TEST DATA
x_test_prediction = model1.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("ACCURACY SCORE ON TEST DATA : ",test_data_accuracy)
```
